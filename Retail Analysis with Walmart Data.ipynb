## Retail Analysis with Walmart Data

```python
#import required libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Understanding the data
walmart=pd.read_csv(r'data\Walmart_Store_sales.csv')
walmart
```

```python
walmart.info()
```

```python
walmart['Date']=pd.to_datetime(walmart['Date'])
walmart.isnull().sum()
```

## Basic Statistics tasks

```python
# Which store has maximum sales
walmart.groupby('Store').Weekly_Sales.sum().sort_values(ascending=False).head(1).round()
```

```python
# Which store has maximum standard deviation i.e., 
#the sales vary a lot. Also, find out the coefficient of mean to standard deviation
walmart.groupby('Store').Weekly_Sales.std().sort_values(ascending=False).round()
```

```python
Store14=walmart[walmart.Store==14]
cof=Store14.Weekly_Sales.std()/Store14.Weekly_Sales.mean()*100
cof
```

```python
# Which store/s has good quarterly growth rate in Q3â€™2012
q2_start=pd.Timestamp(2012,4,1)
q2_stop=pd.Timestamp(2012,6,30)
q3_start=pd.Timestamp(2012,7,1)
q3_stop=pd.Timestamp(2012,9,30)
```

```python
q2walmart=walmart[(walmart['Date']>q2_start) & (walmart['Date']<q2_stop) ]
q3walmart=walmart[(walmart['Date']>q3_start) & (walmart['Date']<q3_stop) ]
```

```python
q2=pd.DataFrame(q2walmart.groupby('Store')['Weekly_Sales'].sum())
q2
```

```python
q3=pd.DataFrame(q3walmart.groupby('Store')['Weekly_Sales'].sum())
q3
```

```python
# Which store/s has good quarterly growth rate in Q3â€™2012
q3_growth=q3.merge(q2, on="Store",how='inner')
q3_growth['growth']=q3_growth['Weekly_Sales_x']-q3_growth['Weekly_Sales_y']
q3_growth['growth_per']=(q3_growth['growth']/q3_growth['Weekly_Sales_x']*100)
q3_growth.sort_values(by=['growth_per'],ascending=False).head(5)
```

## Statistical Model

```python
## For Store 1 â€“ Build  prediction models to forecast demand
data=walmart[walmart['Store']==1]
data
```

```python
## #Change dates into days by creating new variable.

data['Day']=pd.DatetimeIndex(data['Date']).day
data
```

```python
# # Hypothesize if CPI, unemployment, and fuel price have any impact on sales.
import seaborn as sns
d=data.corr()
ax = plt.subplots(figsize=(15, 8))
ax=sns.heatmap(data=d,annot=True,linewidths=.5)
```

```python
## Linear Regression â€“ Utilize variables like date and restructure dates as 1 for 5 Feb 2010 (starting from the earliest date in order). Hypothesize if CPI, unemployment, and fuel price have any impact on sales.
X=data.drop(['Weekly_Sales','Date','Temperature'],axis=1)
y=data['Weekly_Sales']


```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

## sklearn.linear_model.LinearRegressionÂ¶

```python
#Linear Regression â€“ Utilize variables like date and restructure dates as 1 for 5 Feb 2010 
#(starting from the earliest date in order)

from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score
reg = LinearRegression().fit(X_train, y_train)
y_pred=reg.predict(X_test)
print(mean_squared_error(y_test, y_pred))
r2_score(y_test, y_pred)*100
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0)
regr.fit(X_train, y_train)
y_pred=regr.predict(X_test)
print(mean_squared_error(y_test, y_pred))
r2_score(y_test, y_pred)*100
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(X_train, y_train)
y_pred=regr.predict(X_test)
print(mean_squared_error(y_test, y_pred))
r2_score(y_test, y_pred)*100
```

```python
lm = LinearRegression()
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lm, X_train, y_train, scoring='r2', cv=5)
print(scores)
print('Mean Score:' ,scores.mean()*100,'%')
```
